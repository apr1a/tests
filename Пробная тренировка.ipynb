# Общий вывод

**Опции при пробной тренировке:**

- Пробная тренировка 
  - бесплатна
  - со скидкой
  
- При приобретении абонемента/программы в день пробной тренировки - скидка
- Скидка, если привести ещё одного человека
- Стоимость пробного занятия возвращается при покупке абонемента
- Занятие без записи стоит дороже
- Единоразовая возможность покупки абонемента по выгодной цене
- Во всех направлениях для посещения пробной тренировки нужно оставить контакты для связи
