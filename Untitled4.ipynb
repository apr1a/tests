```python
import pandas as pd
df = pd.read_csv('ILSS-Payments-Report-Booking-2022-10.csv')
df = df.loc[df['Region title'].isin(['Москва'])]
df.head()
```

```python
df['Created at'] = pd.to_datetime(df['Created at'])
df['day'] = df['Created at'].dt.day
df
```

```python
def day(row):
    day = row['day']

    if day <=9:
        return 1
    elif 9< day <=16:
        return 2
    elif 16< day <=23:
        return 3
    elif 23< day <=31:
        return 4
df['week'] = df.apply(day, axis=1)
    
df.head() 
```

```python
df = df.loc[df['Region title'].isin(['Москва'])]
df.head()
```

```python
data_pivot = df.pivot_table(index = ['Program title'],
                            columns = ['week'],
                             values = ['Total amount cents'],
                             aggfunc = ['count'])
data_pivot.fillna(0)
```
