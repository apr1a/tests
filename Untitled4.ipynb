```python
import pandas as pd
df = pd.read_csv('ILSS-Payments-Report-Booking-2022-10.csv')
df
```

```python
df['Created at'].astype('int')
```

```python
df = df.loc[df['Region title'].isin(['Москва'])]
df.head()
```

```python
data_pivot = df.pivot_table(index = ['Program title', 'Direction title'],
                             values = ['Total amount cents'],
                             aggfunc = ['count'])
data_pivot
```

```python
data_pivot.plot(kind='bar', title = 'Октябрь', figsize=(16,5))
```
